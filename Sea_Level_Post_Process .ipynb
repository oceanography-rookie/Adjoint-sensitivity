```python
import matplotlib.pyplot as plt
import xarray as xr
import numpy as np
import pandas as pd
plt.rcParams['figure.figsize'] = (15,10)
%matplotlib inline
```

```python
hist = xr.open_dataset('hist_state.0000000000.t001.nc')
rcp = xr.open_dataset('rcp_state.0000000000.t001.nc')
hist.Eta
```

```python
hist.Eta.sel(T='1948-07-26T03:00:00.000000000').plot()
plt.title("free-surface_r-anomaly[m] (1948-07-26T03)")
plt.show()
```

```python
std_hist=hist.Eta.resample(T="1M").mean().std(dim="T")
std_hist.plot()
plt.title("Hist STD")
plt.show()
```

```python
std_rcp=rcp.Eta.resample(T="1M").mean().std(dim="T")
std_rcp.plot()
plt.title("RCP STD")
plt.show()
```

```python
(std_rcp-std_hist).plot()
plt.title("RCP4.5-hist STD")
plt.show()
```
